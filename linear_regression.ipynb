# Part B - Linear Regression

# Reading and Pre-processing Data

```python
import pandas as pd
dataframe=pd.read_csv('FW_Veg_Rem_Combined.csv')
df=dataframe.copy()
df.keys()
```

```python
df.head()
```

```python
import numpy as np
df.dropna(axis='rows', how='any', inplace=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
training_set, testing_set = train_test_split(df, test_size=0.30, random_state=590)
print(training_set.info())
print(testing_set.info())
```

```python
training_set.head()
```

```python
testing_set.head()
```

## 2. Features Selected

I have used the following features for the model:

X = "Temp_pre_7","Wind_pre_7","Hum_pre_7","Prec_pre_7"

Y = "fire_size"

## Reasons

* These values are all float values, hence, computation will be faster (converting String to int or float are time and cost consuming).
* For the purpose of my project: I am planning on predicting the magnitude or size of the fire based on the weather conditions like humidity, wind, temperature and precipitation. The raw data recorded in the dataset that I am using contains these values for the current day of fire, 7 days before the fire, 15 days before the fire and 30 days before the fire.
* Obviously, I cannot use all the feature as it will be overfitting the values to the model. Hence, I am choosing only 4 features, weather conditions from 7 days before the fire.

## 3. Linear Regression


Trainging Set

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
rege = LinearRegression()
X = training_set[["Temp_pre_7","Wind_pre_7","Hum_pre_7","Prec_pre_7"]]
Y = training_set["fire_size"]
print(rege.fit(X, Y))
print('R2 value is : ',rege.score(X,Y))
y_predict = rege.predict(X)
mean_square = mean_squared_error(Y, y_predict)
print("Mean Squared Error is : ",  mean_square)
```

Testing Set

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
rege = LinearRegression()
X = testing_set[["Temp_pre_7","Wind_pre_7","Hum_pre_7","Prec_pre_7"]]
Y = testing_set["fire_size"]
print(rege.fit(X, Y))
print('R2 value is : ',rege.score(X,Y))
y_predict = rege.predict(X)
mean_square = mean_squared_error(Y, y_predict)
print("Mean Squared Error is : ",  mean_square)
```
