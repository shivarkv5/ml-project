# Reading the data

```python
import pandas as pd
dataframe=pd.read_csv('FW_Veg_Rem_Combined.csv')
df=dataframe.copy()
df.info()
```

# Cleaning the data


```python
import numpy as np
df.dropna(axis='rows', how='any', inplace=True)
df.head()
```

# Spliting the data into train set and test set

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.30, random_state=568)
print(train_set.info())
print(test_set.info())
```

# 1. Clustering - K-Means Algorithm


```python
from sklearn.cluster import KMeans
import math

kmeans_cluster = KMeans(init="k-means++", n_clusters=4, n_init=20)

X = train_set[["Temp_pre_7","Wind_pre_7","Hum_pre_7","Prec_pre_7"]]
# Y = train_set["fire_size"]

y = kmeans_cluster.fit(X)

print("Centroids are: ", kmeans_cluster.cluster_centers_)
print("The final distortion is: ", kmeans_cluster.inertia_)

class_out = kmeans_cluster.predict(X)
print("The predicted class: ", class_out)
print("The distortion is ", kmeans_cluster.inertia_)

from sklearn.cluster import KMeans
train_set['x']=train_set['Temp_pre_7']
train_set['x']=train_set['x'].astype(float)
train_set['y']=train_set['fire_size']
train_set['y']=train_set['y'].astype(float)
X = train_set[["x", "y"]]

kmeans_cluster = KMeans(n_clusters=2)
kmeans_cluster.fit(X)

import math
print("The distortion is ", kmeans_cluster.inertia_)

RMS_average = math.sqrt(kmeans_cluster.inertia_ / len(X))
print("Averaging and taking the square root gives ", RMS_average)
print("The centroids are ", kmeans_cluster.cluster_centers_)
print("Assignments: ")
print(kmeans_cluster.labels_)

import numpy as np
import matplotlib.pyplot as plt
plt.scatter(train_set['x'], train_set['y'], color="green")
plt.xlabel("Temp_pre_7")
plt.ylabel("fire_size")
plt.show()


cluster_x = [x for [x,y] in kmeans_cluster.cluster_centers_]
cluster_y = [y for [x,y] in kmeans_cluster.cluster_centers_]
import matplotlib.pyplot as plt
plt.scatter(train_set['x'], train_set['y'], c=kmeans_cluster.labels_)
plt.scatter(cluster_x, cluster_y, marker="+", color="blue")
plt.xlabel("Temp_pre_7")
plt.ylabel("fire_size")
plt.show()
```

# 2. Dimentional Analysis - PSA


```python
import numpy as np
from sklearn.decomposition import PCA

X = train_set[["Temp_pre_7","Wind_pre_7","Hum_pre_7","Prec_pre_7"]]

pca = PCA(n_components=4, svd_solver='full')
y_pred_pca = pca.fit(X)
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```

# Advanced Regression/Classifier Tools:
# 5. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score,recall_score, f1_score, confusion_matrix
from sklearn.metrics import confusion_matrix

# Training Set
X = test_set[["Temp_pre_7","Wind_pre_7","Hum_pre_7","Prec_pre_7"]]
Y = test_set["fire_size_class"]

rfc = RandomForestClassifier()
print(rfc.fit(X,Y))

y_pred = rfc.predict(X)
c_matrix = confusion_matrix(Y, y_pred)
print("Confusion Metrix: ")
print(c_matrix)

print("The Evalution metrics are as follows: ")
print("Accuracy: ", accuracy_score(Y, y_pred))
print("Precision: ", precision_score(Y, y_pred, average="weighted"))
print("Sensitivity: ", recall_score(Y, y_pred, average="weighted"))
print("F1 Score: ", f1_score(Y, y_pred, average="weighted"))
```

# Neural Nets

```python
from sklearn.neural_network import MLPClassifier

# Training Set
X = test_set[["Temp_pre_7","Wind_pre_7","Hum_pre_7","Prec_pre_7"]]
Y = test_set["fire_size_class"]

clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
clf.fit(X, Y)
```

```python
y_pred = clf.predict(X)
c_matrix = confusion_matrix(Y, y_pred)
print("Confusion Metrix: ")
print(c_matrix)

print("The Evalution metrics are as follows: ")
print("Accuracy: ", accuracy_score(Y, y_pred))
print("Precision: ", precision_score(Y, y_pred, average="weighted"))
print("Sensitivity: ", recall_score(Y, y_pred, average="weighted"))
print("F1 Score: ", f1_score(Y, y_pred, average="weighted"))
```
